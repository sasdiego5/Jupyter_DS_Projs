## Storytelling through Data Visualization with Matplotib

* The Department of Education Statistics releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. 
* The data set is broken up into 17 categories of degrees, with each column as a separate category.

* Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. 
* Download the dataset Randal compiled here:
http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
print(women_degrees.head())
women_degrees.describe()

```

## Initial Data Visualization for STEM degrees

```python
# dark blue code for color blind friendly graphs
cb_dark_blue = (0/255,107/255,164/255)
# orange code for color blind friendly graphs
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    # in graph text, x and y axes serve as location coordinates
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

## Visualization diagram across all degrees:
* There are seventeen degrees to generate line charts for.
* Use a subplot grid layout of 6 rows by 3 columns. 
* Then group the degrees into STEM, liberal arts, and other.
* All three lists are ordered in descending order by the mean percentage of degrees awarded to women.

To follow data-ink ratio principles:
* Hide x-axis labels except for the bottom most chart in each column
* Set the y-axis labels to show 0 and 100 only
* Add a horizontal line at the 50% line
* Export into a png file

```python
stem_degs = ["Psychology","Biology","Math and Statistics","Physical Sciences","Computer Science","Engineering"]
lib_art_degs = ["Foreign Languages","English","Communications and Journalism","Art and Performance","Social Sciences and History"]
other_degs = ["Health Professions","Public Administration","Education","Agriculture","Business","Architecture"]
```

```python
# dark blue code for color blind friendly graphs
cb_dark_blue = (0/255,107/255,164/255)
# orange code for color blind friendly graphs
cb_orange = (255/255, 128/255, 14/255)
# gray code for color blind friendly graphs
cb_gray = (171/255, 171/255, 171/255)

fig2 = plt.figure(figsize=(20, 20))

for sp in range(0,18,3):
    # creates indexes 0,1,2,3,4,5 to access corresponding list
    deg_index = int(sp/3)
    #  subplot positions in first column: 1,4,7,10,13,16
    ax = fig2.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_degs[deg_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_degs[deg_index]], c=cb_orange, label='Men', linewidth=3) 
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_degs[deg_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,100])
    # generate a horizontal line through the 50% mark
    # alpha parameter sets the level of transparency bw 0 - 1
    ax.axhline(50, c=cb_gray, alpha=0.5)
    # in graph text, x and y axes serve as location coordinates
    if deg_index == 0:
        ax.text(2002, 80, 'Men')
        ax.text(2002, 10, 'Women')
    elif deg_index == 5:
        ax.text(2005, 85, 'Men')
        ax.text(2005, 10, 'Women')
        ax.tick_params(labelbottom="on")
        
for sp in range(1,16,3):
    # creates indexes: 0,1,2,3,4,5 to access corresponding list
    deg_index = int((sp-1)/3)
    #  subplot positions  in second column: 2,5,8,11,14 
    ax = fig2.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_art_degs[deg_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_art_degs[deg_index]], c=cb_orange, label='Men', linewidth=3) 
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_art_degs[deg_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,100])
    # generate a horizontal line through the 50% mark
    # alpha parameter sets the level of transparency bw 0 - 1
    ax.axhline(50, c=cb_gray, alpha=0.5)
    # in graph text, x and y axes serve as location coordinates
    if deg_index == 0:
        ax.text(2002, 80, 'Men')
        ax.text(2002, 20, 'Women')
    elif deg_index == 4:
        ax.text(2005, 55, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom="on")



for sp in range(2,20,3):
    # creates indexes: 0,1,2,3,4,5 to access corresponding list
    deg_index = int((sp-2)/3)
    #  subplot positions  in third column: 3,6,9,12,15,18
    ax = fig2.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_degs[deg_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_degs[deg_index]], c=cb_orange, label='Men', linewidth=3) 
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_degs[deg_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,100])
    # generate a horizontal line through the 50% mark
    # alpha parameter sets the level of transparency bw 0 - 1
    ax.axhline(50, c=cb_gray, alpha=0.5)
    # in graph text, x and y axes serve as location coordinates
    if deg_index == 0:
        ax.text(2005, 90, 'Men')
        ax.text(2005, 5, 'Women')
    elif deg_index == 5:
        ax.text(2003, 60, 'Men')
        ax.text(2002, 30, 'Women')
        ax.tick_params(labelbottom="on")

# Export file before calling pyplot.show()
plt.savefig("gender_degrees.png")    
plt.show()
 

```
