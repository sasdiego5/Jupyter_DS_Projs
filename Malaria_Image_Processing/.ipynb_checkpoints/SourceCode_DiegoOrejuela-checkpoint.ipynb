## AI Project: Detecting Cells Infected with Malaria

The goal of this project is to create a generalizable model that has the ability to detect an early presence, in human red blood cells, of the parasite that causes human beings to be victims of malaria. 
The Malaria dataset, chosen for this project, has been created by the U.S National Library of Medicine and downloaded through Kaggle. Data consists of 27558 images of human red blood cell samples in an RGB color model.
The average size of the images is approximately 150 by 150. Half of the samples, totaling 13779, are infected with the parasite and the other half are uninfected.
There are three different models implemented from the sequential model in KERAS: a simple MLP form, a basic CNN, and a CNN with data augmentation methods from the ImageDataGenerator class in KERAS.
There are model accuracy and model loss graphs and results after each model, which allow to compare across models and choose the best implementation for a generalizable model.

Dataset Link: [Malaria Cell Images Dataset](https://www.kaggle.com/iarunava/cell-images-for-detecting-malaria)

```python
# import necessary libraries
import tensorflow as tf
import os
import cv2
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import keras
from IPython.display import SVG, display
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, Flatten, Dropout, Conv2D, MaxPooling2D 
from keras.utils import np_utils, plot_model, to_categorical
from keras.utils.vis_utils import model_to_dot
from keras.losses import categorical_crossentropy, binary_crossentropy
from keras.optimizers import SGD
from keras.preprocessing.image import ImageDataGenerator,img_to_array, load_img
from keras.preprocessing import image
from keras.backend.tensorflow_backend import set_session

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)
set_session(sess)
```

```python
#Set NumPy and TensorFlow seeds for reproducible results
from numpy.random import seed
from tensorflow import set_random_seed

seed(764)
set_random_seed(764)
```

```python
# create a list of items in the directory
# remove unnecessary files
print(os.listdir('cell_images'))
parasitized = os.listdir('cell_images/Parasitized/')
parasitized.remove('.DS_Store')
uninfected = os.listdir('cell_images/Uninfected/')
uninfected.remove('.DS_Store')
print("Parasitized images: ",len(parasitized))
print("Uninfected images: ",len(uninfected))
```

```python
# create an empty list to append both datasets together
dataset = []

#for loop to append corresponding label to the file images
for i in parasitized:
    dataset.append(['cell_images/Parasitized/'+i,0])
for i in uninfected:
    dataset.append(['cell_images/Uninfected/'+i,1])
print("Total Number of Images: ",len(dataset))
```

```python
# Visualize Images both types of images

plt.figure(figsize=(20,20))
for i in range(10):
    plt.subplot(1,10,i+1)
    img = load_img('cell_images/Parasitized/'+ parasitized[i])
    plt.imshow(img)
    plt.title('infected')
    
plt.figure(figsize=(20,20))
for i in range(10):
    plt.subplot(1,10,i+1)
    img2 = load_img('cell_images/Uninfected/'+ uninfected[i])
    plt.imshow(img2)
    plt.title('uninfected')
```

```python
# shuffle dataset, split images from labels, create train and test datasets
np.random.shuffle(dataset)

cells=[]
for i in dataset:
    cells.append(i[0])
    
labels=[]
for i in dataset:
    labels.append(i[1])

X_train, X_test, Y_train, Y_test = train_test_split(cells, labels, test_size = 0.2)
```

# Model 1: MLP

```python
# function to pre process images to use as input in MLP
def prepare_mlp_imgs(img_dir):
    img = load_img(img_dir,grayscale=True)
    img = img.resize((64,64))
    img = img_to_array(img)
    img = img.reshape(64,64)
    return img
```

```python
x_train_mlp  = []
y_train_mlp  = []
x_test_mlp  = []
y_test_mlp  = []

for i in range(len(X_train)):
        x_train_mlp.append(prepare_mlp_imgs(X_train[i]))
        y_train_mlp.append(Y_train[i])
        
for i in range(len(X_test)):
        x_test_mlp.append(prepare_mlp_imgs(X_test[i]))
        y_test_mlp.append(Y_test[i])
        
x_train_mlp  = np.array(x_train_mlp )
y_train_mlp   = np.array(y_train_mlp )
x_test_mlp   = np.array(x_test_mlp )
y_test_mlp   = np.array(y_test_mlp )

print('\nytrain:')
print(y_train_mlp .shape)

print('\nytest:')
print(y_test_mlp .shape)

print('\nxtrain:')
print(x_train_mlp .shape)

print('\nxtest:')
print(x_test_mlp .shape)
```

```python
# Data Preprocessing
xsize = (4096,)

#Normalize image bytes (0-255) to a scale of 0-1
x_train_mlp  = x_train_mlp  / 255
x_test_mlp  = x_test_mlp  / 255

#Reshape x for a multi-layer perceptron (channels last)
x_train_mlp  = x_train_mlp .reshape(x_train_mlp.shape[0], xsize[0])
x_test_mlp  = x_test_mlp .reshape(x_test_mlp.shape[0], xsize[0])

#Double check shape
print('\nytrain:')
print(y_train_mlp .shape)

print('\nytest:')
print(y_test_mlp .shape)

print('\nxtrain:')
print(x_train_mlp .shape)

print('\nxtest:')
print(x_test_mlp .shape)
```

```python
# Create and Train the model 
#Model configurations
epochs = 30
batchsize = 128

learning_rate = .01

#Create your multi-layer perceptron
mlp = Sequential()

#Start with a 64-node dense layer that has 4096 inputs (flattened 64 * 64 image)
mlp.add(Dense(128, activation='relu', input_shape=(4096,)))
# mlp.add(Dropout(0.1))

#Followed by a 32-node dense layer with relu activation
mlp.add(Dense(64, activation='relu'))

#Followed by a 32-node dense layer with relu activation
mlp.add(Dense(32, activation='relu'))

#Followed by a 16-node dense layer with relu activation
mlp.add(Dense(16, activation='relu'))
    
#Finally, output a propability between 0 and 1 for each class
mlp.add(Dense(1, activation='sigmoid'))
mlp.compile(loss=binary_crossentropy,optimizer=SGD(lr=learning_rate), metrics=['accuracy'])
# optimizer="adam"

trained = mlp.fit(x_train_mlp  , y_train_mlp , batch_size=batchsize,epochs=epochs, 
                  verbose=1, validation_data=(x_test_mlp  , y_test_mlp ))
```

```python
#Visualize what was ran, start with model.summary() and pydot.Graph:
print(mlp.summary())
display(SVG(model_to_dot(mlp).create(prog='dot',format='svg')))
```

```python
#Show model performance using code from Keras site: https://keras.io/visualization/
%matplotlib inline
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(trained.history['acc'])
plt.plot(trained.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(trained.history['loss'])
plt.plot(trained.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
mlp_results = mlp.evaluate(x_test_mlp,y_test_mlp,verbose=1)
print("Test set loss:",mlp_results[0])
print("Test set accuracy:",mlp_results[1])
```

# Model 2: CNN for RGB images

```python
def prepare_cnn_imgs(img_dir):
    img = load_img(img_dir)
    img = img.resize((64,64))
    img = img_to_array(img)
    return img
```

```python
x_train_m2 = []
y_train_m2 = []
x_test_m2 = []
y_test_m2 = []

for i in range(len(X_train)):
        x_train_m2.append(prepare_cnn_imgs(X_train[i]))
        y_train_m2.append(Y_train[i])
        
for i in range(len(X_test)):
        x_test_m2.append(prepare_cnn_imgs(X_test[i]))
        y_test_m2.append(Y_test[i])

x_train_m2 = np.array(x_train_m2)
y_train_m2 = np.array(y_train_m2)
x_test_m2 = np.array(x_test_m2)
y_test_m2 = np.array(y_test_m2)

print('\nxtrain:')
print(x_train_m2.shape)

print('\nytrain:')
print(y_train_m2.shape)

print('\nxtest:')
print(x_test_m2.shape)

print('\nytest:')
print(y_test_m2.shape)

#Normalize image bytes (0-255) to a scale of 0-1
x_train_m2  = x_train_m2 / 255
x_test_m2 = x_test_m2 / 255
```

```python
# build cnn model 
batchsize = 32
epochs = 10

model = Sequential()

model.add(Conv2D(64, kernel_size=(4,4), strides =1 ,  activation='relu', input_shape=x_train_m2.shape[1:]))
model.add(Conv2D(32, kernel_size=(4,4),strides =1 , activation='relu'))
model.add(Conv2D(16, kernel_size=(4,4),strides =1 , activation='relu'))
model.add(Flatten())
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#train cnn model
trained = model.fit(x_train_m2, y_train_m2, batch_size=batchsize, epochs=epochs, 
                    verbose=1, validation_data=(x_test_m2, y_test_m2))
```

```python
#Visualize what was ran, start with model.summary() and pydot.Graph:
print(model.summary())
display(SVG(model_to_dot(model).create(prog='dot',format='svg')))
```

```python
#Show model performance using code from Keras site: https://keras.io/visualization/
%matplotlib inline
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(trained.history['acc'])
plt.plot(trained.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(trained.history['loss'])
plt.plot(trained.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
model_results = model.evaluate(x_test_m2,y_test_m2,verbose=1)
print("Test set loss:",model_results[0])
print("Test set accuracy:",model_results[1])
```

# Model 3: Deep CNN for RGB images with Data Augmentation

```python
# implement Keras ImageDataGenerator object to modify training and test sets
# use some of transformations available for data augmentation

y_train_m3 = to_categorical(y_train_m2,num_classes=2)
y_test_m3 = to_categorical(y_test_m2,num_classes=2)

train_datagen = ImageDataGenerator(rotation_range = 25, width_shift_range = 0.25, height_shift_range = 0.25,
                                     shear_range = 0.25 ,zoom_range = 0.25 , vertical_flip = True)
train_generator = train_datagen.flow(x_train_m2,y_train_m3, batch_size=64, shuffle=False)

test_datagen = ImageDataGenerator()

validation_generator = test_datagen.flow(x_test_m2,y_test_m3, batch_size=64,shuffle=False)
```

```python
batchsize = 64
epochs = 32
learning_rate = .01

# create four convolutions with their respective activations
model_2 = Sequential()
c1 = Conv2D(filters=32, kernel_size = (4, 4), strides=1, activation='relu', input_shape=x_train_m2.shape[1:])
c2 = Conv2D(filters=32, kernel_size = (4, 4), strides=1, activation='relu')
c3 = Conv2D(filters=64, kernel_size = (3, 3), strides=1, activation='relu')
c4 = Conv2D(filters=64, kernel_size = (3, 3), strides=1, activation='relu')

# add convolutions and pooling to the model
model_2.add(c1)
model_2.add(c2)
model_2.add(MaxPooling2D(pool_size=(2,2)))
model_2.add(BatchNormalization())
model_2.add(Dropout(.2))
model_2.add(c3)
model_2.add(MaxPooling2D(pool_size=(2,2)))
model_2.add(BatchNormalization())
model_2.add(Dropout(.2))
model_2.add(c4)
model_2.add(MaxPooling2D(pool_size=(2,2)))
model_2.add(BatchNormalization())
model_2.add(Dropout(.2))
model_2.add(Flatten())

# add fully connected layer for classification
model_2.add(Dense(128, activation='relu'))

# last layer with number of outputs equal to number of classes
model_2.add(Dense(2, activation='sigmoid'))

# compile the model and predict classifications
op = SGD(learning_rate)
model_2.compile(optimizer=op, loss='categorical_crossentropy', metrics=['accuracy'])

trained_aug = model_2.fit_generator(train_generator, steps_per_epoch = len(x_train_m2)//batchsize, 
                                    epochs = epochs, validation_data = validation_generator, 
                                    validation_steps = len(x_test_m2)//batchsize, verbose = 1)   
```

```python
#Visualize what was ran, start with model.summary()
print(model_2.summary())
display(SVG(model_to_dot(model_2).create(prog='dot',format='svg')))
```

```python
#Show model performance using code from Keras site: https://keras.io/visualization/
%matplotlib inline
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(trained_aug.history['acc'])
plt.plot(trained_aug.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(trained_aug.history['loss'])
plt.plot(trained_aug.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
model_2_results = model_2.evaluate_generator(validation_generator, verbose=1,steps=8)

print("Test set loss:",model_2_results[0])
print("Test set accuracy:",model_2_results[1])
```
