## Decision Tree Graph for main causes of employee attrition on employee dataset

The goal of this project is to implement a decision tree based on a dataset that contains information about employees who left or stayed at a certain company and the factors that could affect their decision.
At the end of the notebook a Random Forest Classifier is implemented to test for better prediction results with the use fo a confussion matrix.

```python
# import required libaries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#load dataset
dataset = pd.read_excel('employee.xlsx')
```

```python
dataset.dtypes
```

```python
dataset.head()
```

```python
#col 1 is not needed, slice this way for X to be a matrix
X = dataset.iloc[: , [0,1,2,3,4,5,7,8,9]].values
y = dataset.iloc[ : , 6].values

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelEncoder_X = LabelEncoder()
X[: , -2] = labelEncoder_X.fit_transform(X[ : , -2])
X[: , -1] = labelEncoder_X.fit_transform(X[ : , -1])

#split data into Training and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)


#Normalizing the features
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

# Make Classifier object called classifierObj
from sklearn.tree  import DecisionTreeClassifier
classifierObj = DecisionTreeClassifier(criterion='gini', max_depth =4, random_state=1)
classifierObj.fit(X_train, y_train)

#Making Predictions on the Test Set
y_pred = classifierObj.predict(X_test)

#Evaluate predictions using a Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz

dot_data = export_graphviz(classifierObj, filled=True, rounded=True, class_names=['stay', 'left'],
                          feature_names=['satisfaction_level','last_evaluation','number_project','average_monthly_hours','time_spend_company','work_accident','promotion_last_5years','department','salary'],
                           out_file=None)
graph = graph_from_dot_data(dot_data)
graph.write_png('tree.png')
graph.write_jpeg('tree.jpeg')
```

```python
# implementing a random forest classifier
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion="gini", n_estimators=25, random_state=1,n_jobs=2)
forest.fit(X_train,y_train)
#Making Predictions on the Test Set
forest_y_pred = forest.predict(X_test)

#Evaluate predictions using a Confusion Matrix
forest_cm = confusion_matrix(y_test, forest_y_pred)
print(forest_cm)
```
