## Profitable Android and iOS app types

The goal of this project is to analyze data in order to understand what kinds of apps are likely to attract more users. 
There will be analysis made for both Android and iOS apps available on Google Play and the App Store.

The datasets are available on kaggle, the links below provide more information on each dataset:

* [google play store](https://www.kaggle.com/lava18/google-play-store-apps/home)
* [apple store](https://www.kaggle.com/ramamet4/app-store-apple-data-set-10k-apps/home)

```python
# function that enables to print rows from data in a more readable way
def explore_data(dataset, start, end, rows_cols=False):
    dataset_slice = dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n')
        
    if rows_cols:
        print("Number of rows: ", len(dataset))
        print("Number of columns: ", len(dataset[0]))
```

```python
from csv import reader

# Google Play data set 
opened_file = open('googleplaystore.csv')
read_file = reader(opened_file)
google = list(read_file)
google_header = google[0]
google = google[1:]

# App Store data set 
opened_file = open('AppleStore.csv')
read_file = reader(opened_file)
apple = list(read_file)
apple_header = apple[0]
apple = apple[1:]
```

```python
print(google_header)
print('\n')
explore_data(google,0,5,True)
```

```python
print(apple_header)
print("\n")
explore_data(apple,0,5,True)
```

## Deleting duplicate entries 
* first find the duplicate entries
* then analyze why this entries are duplicated
* Not removing a random duplicated row
* Instead evaluate duplicates based on the highest number of reviews
* the more reviews suggest more reliability in the ratings column

```python
google_duplicate_apps = []
google_unique_apps = []

for app in google:
    name = app[0]
    if name in google_unique_apps:
        google_duplicate_apps.append(name)
    google_unique_apps.append(name)
    
print(len(google_duplicate_apps))
print(google_duplicate_apps[:15])
```

```python
apple_duplicate_apps = []
apple_unique_apps = []

for app in apple:
    name = app[0]
    if name in apple_unique_apps:
        apple_duplicate_apps.append(name)
    apple_unique_apps.append(name)
    
print(len(apple_duplicate_apps))
print(apple_duplicate_apps[:15])
```

Steps to remove the duplicates:

* Create a dictionary, where each dictionary key is a unique app name and the corresponding dictionary value is the highest number of reviews of that app

* Use the information stored in the dictionary to create a new data set, which will have only one entry per app (and for each app, only select the entry with the highest number of reviews)

```python
reviews_max = {}
for app in google:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name]= n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
print(len(reviews_max))
```

```python
google_clean=[]
already_added=[]
for app in google:
    name = app[0]
    n_reviews = float(app[3])
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        google_clean.append(app)
        already_added.append(name)
explore_data(google_clean,0,5,True)
```

## Removing apps that are not in English
* Write a function that takes in a string and returns False if there's any character in the string that doesn't belong to the set of common English characters, otherwise it returns True.
* English letters range from 0-127 in the ASCII system
* To minimize data loss, remove an app only if it has more than three non-ASCII characters

```python
def english_letters(string):
    non_ascii = 0
    for char in string:
        if ord(char) > 127:
            non_ascii +=1
    if non_ascii > 3:
            return False
    else:
        return True
# test the function
print(english_letters('Instagram'))
print(english_letters('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(english_letters('Docs To Go™ Free Office Suite'))
print(english_letters('Instachat 😜'))
```

```python
# use function to create two new lists of apple store and google play store apps that are only in english
apple_in_english=[]
google_in_english=[]

for app in google_clean:
    name = app[0]
    if english_letters(name):
        google_in_english.append(app)
for app in apple:
    name=app[1]
    if english_letters(name):
        apple_in_english.append(app)

explore_data(google_in_english,0,5,True)
print("\n")
explore_data(apple_in_english,0,5,True)
    
```

## Most popular apps by genre
* build frequency tables for relevant columns from both datasets

* find most common genres in each market
    
        ios: prime_genre and for android: genres and category
    
    
* build two functions to analyze the frequency tables:

        One function to generate frequency tables that show percentages
    
        Another function to display the percentages in a descending order

```python
# rename clean data lists ready for analysis
ios = apple_free.copy()
android = google_free.copy()
```

```python
# function for generating frequency tables
def freq_table(dataset, index):
    
    freq_dict ={}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in freq_dict:
            freq_dict[value]+=1
        else:
            freq_dict[value]=1
        
    freq_percentages = {}
    for key in freq_dict:
        percentage = (freq_dict[key] / total) * 100
        freq_percentages[key] = percentage 
 
    return freq_percentages

# displays frequency tables in descending order as a list of tuples
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
# frequency table for app store prime_genre column
display_table(ios,11)
```

```python
# frequency table for google play store category column
display_table(android,1)
```

```python
# frequency table for googgle play store genre column
display_table(android,9)
```

## Most number of users by app genre
* For the Google Play dataset the Installs column provides this information
* The App Store dataset is missing this info. But, the total number of user ratings in the rating_count_tot app may be useful too
* Isolate the apps of each genre
* Sum up the user ratings for the apps of that genre
* Divide the sum by the number of apps belonging to that genre 




```python
# Apple dataset
apple_genres = freq_table(ios,11)

for genre in apple_genres:
    total = 0
    len_genre = 0
    
    for app in ios:
        genre_app = app[11]
        if genre_app == genre:
            ratings_num = float(app[5])
            total += ratings_num
            len_genre +=1

    ratings_average = total / len_genre
    print(genre,":",ratings_average)
```

```python
for app in ios:
    if app[11] == 'Shopping':
        print(app[1], ':', app[5]) # print name and number of ratings
```

```python
# Google Play Store dataset
google_categories = freq_table(android,1)

for category in google_categories:
    total = 0
    len_category = 0
    
    for app in android:
        category_app = app[1]
        if category_app == category:
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```
