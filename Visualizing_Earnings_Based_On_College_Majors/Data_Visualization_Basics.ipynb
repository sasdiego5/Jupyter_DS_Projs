## Data Visualization of earning based on College Majors

* Rank - Rank by median earnings (the dataset is ordered by this column).

* Major_code - Major code.
* Major - Major description.
* Major_category - Category of major.
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time.
* Men - Male graduates.
* Women - Female graduates.
* ShareWomen - Women as share of total.
* Employed - Number employed.
* Median - Median salary of full-time, year-round workers.
* Low_wage_jobs - Number in low-wage service jobs.
* Full_time - Number employed 35 hours or more.
* Part_time - Number employed less than 35 hours.

```python
import pandas as pd
import matplotlib.pyplot as plt 
# Jupyter magic to display plots inline
%matplotlib inline 
```

## Initial data preparation before visualization

```python
recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.shape)
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.isna().sum())
```

```python
# it is valid to delete the null values since there are only four in total
recent_grads = recent_grads.dropna()
```

```python
print(recent_grads.shape)
print(recent_grads.describe())
```

## Visualizing with Scatter Plots

```python
# Sample_size vs Median
recent_grads.plot(x="Sample_size", y='Median',kind='scatter',title='Sample_size vs Median')
```

* The above scatter plot shows that most of the sample_size values of 0-400 have a median in between 20000-60000
* There is no correlation

```python
# Sample_size vs Unemployment_rate
recent_grads.plot(x="Sample_size", y='Unemployment_rate',kind='scatter',title='Sample_size vs Unemployment_Rate')
```

* Most of the values seem to be around sample_size = 0 and Unemployment_rate range 0 - 0.125
* There is no correlation

```python
# Full_time vs Median
recent_grads.plot(x="Full_time", y='Median',kind='scatter',title='Full_time vs Median')
```

* Most of the values seem to be around full_time = 0-50000 and Median = 20000 - 60000
* There is no correlation

```python
# ShareWomen vs Unemployment_rate
recent_grads.plot(x="ShareWomen", y='Unemployment_rate',kind='scatter',title='ShareWomen vs Unemployment_rate')
```

* There seems to be a weak positive correlation

```python
# Men vs Median
recent_grads.plot(x="Men", y="Median",kind='scatter',title='Men vs Median')
```

* Most of the values are for Men = 0-25000 with a Median = 20000-60000
* There is no correlation

```python
# Women vs Median
recent_grads.plot(x="Women", y='Median',kind='scatter',title='Women vs Median')
```

* Most of the values are for Women = 0-50000 with a Median = 20000-60000
* There is no correlation

## Bar plots

```python
# compare first ten rows vs last ten rows of dataframe
recent_grads[:10].plot.bar(x="Major",y="ShareWomen",legend=False,title="Major vs ShareWomen, first 10")
recent_grads[-10:].plot.bar(x="Major",y="ShareWomen",legend=False,title="Major vs ShareWomen, last 10")
```

```python
# compare first ten rows vs last ten rows of dataframe
recent_grads[:10].plot(x="Major",y="Unemployment_rate",legend=False,kind="barh",title="Unemployment_rate vs recent_grad, first 10")
recent_grads[-10:].plot(x="Major",y="Unemployment_rate",legend=False,kind="barh",title="Unemployment_rate vs recent_grad, last 10")
```

## Boxplot

```python
# explore the distributions of median salaries
fig, ax = plt.subplots()
ax.boxplot(recent_grads["Median"].values)
ax.set_title("Distribution of Median")
```

```python
# explore the distributions of unemployment rates
fig, ax = plt.subplots()
ax.boxplot(recent_grads["Unemployment_rate"].values)
ax.set_title("Unemployment_rate")
```

## Use a grouped bar plot to compare the number of men with the number of women in each category of majors.



```python
df_major = recent_grads.set_index(recent_grads["Major"].unique())

df_m_w = pd.DataFrame(df_major[['Men', 'Women']], columns=['Men', 'Women'])

for row in range(0,4):
    start_index= row*43
    end_index=(row+1)*43
    df_m_w[start_index:end_index].plot.bar(figsize=(10,5))
    
```
